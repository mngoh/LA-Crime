# Groupby Data

```python
# Create the grouped bar plot
def group_bar(x_var, y_var):
    sns.barplot(data=df,x=x_var, y=y_var, palette='cool')
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.title(f'Interaction of {x_var} and {y_var}')
    plt.show()
```

```python
# Plot 
group_bar('vict_race', 'vict_age')
```

```python
# Perform value counts on the 'weapon_def' column
weapon_counts = df['weapon_def'].value_counts()

# Filter the DataFrame based on counts greater than 50
weapon = df[df['weapon_def'].isin(weapon_counts[weapon_counts > 1000].index)]
```

```python
# Cross-tabs Weapon Code used
cross_tab = pd.crosstab(weapon['weapon_def'], [weapon['vict_race'], weapon['vict_sex']])
print(cross_tab)
```

### Notes on the Below, Black women are facing ecexptionally larger assualts than black men in area 3, 12*

```python
# Cross-tabs Weapon Code used
cross_tab = pd.crosstab(weapon['area_name'], [weapon['vict_race'], weapon['vict_sex']])
print(cross_tab["B"].sort_values(by = 'F', ascending = False))
```

```python
# Create Data Frame of Areas and plot the % difference
black_area = pd.DataFrame(cross_tab["B"].sort_values(by = 'F', ascending = False))
black_area[:1]
```

```python
# Visualize
black_area['diff'] = round(-((black_area['M']-black_area['F'])/black_area['F'])*100, 2)
black_area[:2]
```

# Plot and Compare 

```python
# Plot the bars
plt.figure(figsize=(10, 5)) 
black_area['diff'].sort_values(ascending = False).plot(kind = 'bar')
plt.title("Areas With The Highest Assault on Black Women vs Black Men", fontsize=16)
plt.suptitle('Values represent the % more crime on Women Vs Men', fontsize=12)
plt.ylabel('Percent Difference', fontsize=15)
plt.xlabel("Area", fontsize=15)
plt.tick_params(axis='x', rotation=45)
plt.show()
```

```python
# Function to View the Variance in victims while considering race
def race(race=''):
    cross_tab = pd.crosstab(weapon['area_name'], [weapon['vict_race'], weapon['vict_sex']])
    black_area = pd.DataFrame(cross_tab[race].sort_values(by='F', ascending=False))
    black_area[:1]
    black_area['diff'] = round(-((black_area['M'] - black_area['F']) / black_area['F']) * 100, 2)
    black_area = black_area.sort_values(by='diff', ascending=False)  # Sort the DataFrame by 'diff' column
    black_area[:2]

    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 5))
    sns.barplot(data=black_area, x=black_area.index, y='diff', color="#3F8EAA" )
    plt.title(f"Areas With The Highest Assault on {race} Women vs {race} Men", fontsize=16)
    plt.suptitle('Values represent the % more crime on Women Vs Men', fontsize=12)
    plt.ylabel('Percent Difference', fontsize=15)
    plt.xlabel("Area", fontsize=15)
    plt.xticks(rotation=45)
    plt.show()

races = ["B", "W", "H", "A"]
for x in races:
    race(race=x)
```
